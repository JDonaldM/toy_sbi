```python
import numpy as np
import matplotlib.pyplot as plt
import getdist
from getdist import plots, MCSamples
```

```python
labels = [r"m",r"c"]
names = ["m","c"]
priors = {"c": [-10, 10]}
```

```python
NDE_chain = np.load("NPE_samples.npy")
NDE_samples = MCSamples(samples=NDE_chain[:,:], names = names, labels = labels)
NDE_samples.updateSettings({'smooth_scale_2D': 0.4, 'smooth_scale_1D':0.2})
```

```python
MCMC_chain = np.load("MCMC_samples.npy")
MCMC_samples = MCSamples(samples=MCMC_chain, names = names, labels = labels)
MCMC_samples.updateSettings({'smooth_scale_2D': 0.4, 'smooth_scale_1D':0.2})
```

```python
GP_chain = np.load("GP_samples.npy")
GP_samples = MCSamples(samples=GP_chain, names = names, labels = labels)
GP_samples.updateSettings({'smooth_scale_2D': 0.4, 'smooth_scale_1D':0.2})
```

```python
g = plots.get_subplot_plotter(width_inch=6)
g.triangle_plot([MCMC_samples, NDE_samples, GP_samples],
                filled=[True, False, False],
                contour_colors=['tab:blue', 'tab:green', 'tab:orange'],
                contour_lws=[2., 2., 2.], 
                legend_labels=['MCMC', 'NPE', 'GP likelihood'], 
                contour_ls=['-', '-', '--'])
```
